#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

x = np.arange(0,100)
y1 = x**2
y2 = math.e**x

fig3, ax3 = plt.subplots(1,2,figsize = [16,5])

ax3[0].set_title("Y = x**2", fontsize = 20)
ax3[1].set_title("Y = e**x", fontsize = 20)

ax3[0].plot(x,y1)
ax3[0].set_xscale("linear")
ax3[0].set_yscale("linear")

ax3[1].plot(x,y2)
ax3[1].set_xscale("log")
ax3[1].set_yscale("log")
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here

x = np.arange(0,100)
y1 = x**2
y2 = math.e**x

fig3, ax3 = plt.subplots(1,2,figsize = [16,5])

ax3[0].set_title("Y = x**2", fontsize = 20)
ax3[1].set_title("Y = e**x", fontsize = 20)

ax3[0].plot(x,y1, label = "Y = x**2")
ax3[0].set_xscale("linear")
ax3[0].set_yscale("linear")

ax3[1].plot(x,y2, label = "Y = e**x", color = "red")
ax3[1].set_xscale("log")
ax3[1].set_yscale("log")

ax3[0].legend()
ax3[1].legend()
```

```python
fitbit_steps_months_work = fitbit[["Steps","Months_encoded","Work_or_Weekend"]]

fitbit_steps_months_work
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

median_steps_month = fitbit_steps_months_work.groupby("Months_encoded")["Steps"].median().reset_index()

median_steps_month
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

median_steps_work = fitbit_steps_months_work.groupby("Work_or_Weekend")["Steps"].median().reset_index()

median_steps_work
```

```python
x1 = median_steps_month["Months_encoded"]
y1 = median_steps_month["Steps"]

x2 = median_steps_work["Work_or_Weekend"]
y2 = median_steps_work["Steps"]

fig4, ax4 = plt.subplots(1,2,figsize = [12,4])

ax4[0].set_title("Months / Steps", fontsize = 20)
ax4[1].set_title("Work or Weekend / Steps", fontsize = 20)

ax4[0].bar(x1,y1, label = "Months / Steps")
ax4[0].set_ylim([7500,14000])

ax4[1].bar(x2,y2, label = "Work or Weekend / Steps", color = "red")
ax4[1].set_ylim([9800,10600])


ax4[0].legend()
ax4[1].legend()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here

bins = [10,20,50]

fig6, ax6 = plt.subplots(1,3,figsize = [16,4])


for x in range(0,3):
    plt.sca(ax6[x])
    titanic["Age"].hist(bins = bins[x])
    plt.xlabel("Ages")
    plt.ylabel(bins[x])
```

#### How does the bin size affect your plot?

```python
"""
The bins are like the separators between values. If we have more bins we can observe with more detail each value on axis X.

For example: In the first subplot we have 10 bins, the max value of Ages is 80. 80 / 10 = 8, each bar have 8 values of year
(first bar: Ages between 0 and 8 years, second bar: between 9 and 17 years, etc)

In the last subplot we have 50 bins, this means that each bar have 1.6 years respect 80.


"""
```
